# Real data!


The following is a way of reading in real weather station data from January 1 to December 31, 2022.

Note that after next week you will probably never not use `pandas` to do this, so bear with me (hah, pandas, bear...)

```python
import numpy as np

precipitation = np.genfromtxt('williamsburg_meteo.csv', delimiter=',', usecols=[4], skip_header=True)
tmax = np.genfromtxt('williamsburg_meteo.csv', delimiter=',', usecols=[5], skip_header=True)
tmin = np.genfromtxt('williamsburg_meteo.csv', delimiter=',', usecols=[6], skip_header=True)
```

Your in-class assignment will constitute plotting these data and creating <b>deliverables</b> from these data.



```python
#First bullet
import matplotlib.pyplot as plt
day = np.arange(1, 366)
tmin_arr = np.array(tmin)
tmax_arr = np.array(tmax)

tmin_c = (tmin - 32) * (5/9)
tmax_c = (tmax - 32) * (5/9)

#sharey=False means they have separate y axes, see set_ylim
fig, ax = plt.subplots(1,3, figsize=(18,6), sharey=False)
ax[0].plot(day, tmax_c, color='red', marker='o', linestyle='-', markersize=2, label= 'Maximum Temperature')
ax[0].plot(day, tmin_c, color='blue', marker='o', linestyle='-', markersize=2, label= 'Minimum Temperature')
#legend uses the labels set above
ax[0].legend()

ax[0].set_title('Minimum and Maximum Temperature Over Time')
ax[0].set_xlabel('Time (days)')
ax[0].set_ylabel('Temperature (C)')
ax[0].set_ylim(-20,40)

#Third bullet 
tdiff_c = (tmax_c - tmin_c)
ax[1].scatter(tdiff_c, precipitation)

ax[1].set_title('Temperature Difference vs. Precipitation')
ax[1].set_ylabel('Precipiation (inches)')
ax[1].set_xlabel('Difference between Maximum and Minimum Temperatures (C)')
ax[1].set_ylim(0,3)

#Fourth bullet
#https://stackoverflow.com/questions/6771428/most-efficient-way-to-reverse-a-numpy-array used to help reverse so 
#rainiest days printed in descending order of highest temperatures.

precip = np.array(precipitation)
precip_sort = np.argsort(precip)
top5_precip = precip_sort[-5:]
descend_top5 = top5_precip[::-1]

print('Maximum temperatures of top 5 rainiest days, with precipitation descending:', tmax_arr[descend_top5])
print('Minimum temperatures of top 5 rainiest days, with precipitation descending:', tmin_arr[descend_top5])

#Second bullet 

tavg_c = (tmax_c + tmin_c)/2
ax[2].plot(day, tmax_c, color= 'red', marker='o', linestyle='-', markersize=2)
ax[2].scatter(day, tavg_c, c=precipitation)
fig.colorbar(label="percent humidity")
```
